```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
result_baseline = pd.read_csv('evaluation_baseline_SVM.csv', usecols=['recall_bs_tr','recall_bs_ts','F1_bs_tr','F1_bs_ts'])
result_entropy = pd.read_csv('evaluation_entropy_SVM.csv', usecols=['recall_bs_tr','recall_bs_ts','F1_bs_tr','F1_bs_ts'])
```

```python
result_baseline_f1_tr = pd.DataFrame(result_baseline['F1_bs_tr'])
result_baseline_f1_tr.columns = ['value']
result_baseline_f1_tr['group'] = 'F1_tr'

result_baseline_f1_ts = pd.DataFrame(result_baseline['F1_bs_ts'])
result_baseline_f1_ts.columns = ['value']
result_baseline_f1_ts['group'] = 'F1_ts'

result_baseline_recall_tr = pd.DataFrame(result_baseline['recall_bs_tr'])
result_baseline_recall_tr.columns = ['value']
result_baseline_recall_tr['group'] = 'recall_tr'

result_baseline_recall_ts = pd.DataFrame(result_baseline['recall_bs_ts'])
result_baseline_recall_ts.columns = ['value']
result_baseline_recall_ts['group'] = 'recall_ts'

result_baseline_n = pd.concat([result_baseline_f1_tr, result_baseline_f1_ts])
result_baseline_n = pd.concat([result_baseline_n, result_baseline_recall_tr])
result_baseline_n = pd.concat([result_baseline_n, result_baseline_recall_ts]).reset_index(drop=True)
result_baseline_n['Features'] = 'baseline'
result_baseline_n
```

```python
result_entropy_f1_tr = pd.DataFrame(result_entropy['F1_bs_tr'])
result_entropy_f1_tr.columns = ['value']
result_entropy_f1_tr['group'] = 'F1_tr'

result_entropy_f1_ts = pd.DataFrame(result_entropy['F1_bs_ts'])
result_entropy_f1_ts.columns = ['value']
result_entropy_f1_ts['group'] = 'F1_ts'

result_baseline_entropy_tr = pd.DataFrame(result_entropy['recall_bs_tr'])
result_baseline_entropy_tr.columns = ['value']
result_baseline_entropy_tr['group'] = 'recall_tr'

result_baseline_entropy_ts = pd.DataFrame(result_entropy['recall_bs_ts'])
result_baseline_entropy_ts.columns = ['value']
result_baseline_entropy_ts['group'] = 'recall_ts'

result_entropy_n = pd.concat([result_entropy_f1_tr, result_entropy_f1_ts])
result_entropy_n = pd.concat([result_entropy_n, result_baseline_entropy_tr])
result_entropy_n = pd.concat([result_entropy_n, result_baseline_entropy_ts]).reset_index(drop=True)
result_entropy_n['Features'] = 'entropy'
result_entropy_n
```

```python
result_all = pd.concat([result_baseline_n, result_entropy_n], axis=0)
# result_all.drop(['group_2'],axis=1, inpl÷ce=True)
result_all
```

```python
ax = sns.boxplot(x="group", y="value", hue="Features",data=result_all, palette="Set3")
# plt.ylim(0,1)
plt.savefig('plot_result_MLP.png', dpi=100)
```
