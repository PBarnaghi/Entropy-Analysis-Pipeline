```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
baseline_MLP = pd.read_csv('auc_baseline_MLP.csv')
baseline_LSTM = pd.read_csv('auc_baseline_LSTM.csv')
entropy_MLP = pd.read_csv('auc_entropy_MLP.csv')
entropy_LSTM = pd.read_csv('auc_entropy_LSTM.csv')
```

```python
baseline_MLP_fpr = baseline_MLP.fpr
baseline_MLP_tpr = baseline_MLP.tpr

baseline_LSTM_fpr = baseline_LSTM.fpr
baseline_LSTM_tpr = baseline_LSTM.tpr

entropy_MLP_fpr = entropy_MLP.fpr
entropy_MLP_tpr = entropy_MLP.tpr

entropy_LSTM_fpr = entropy_LSTM.fpr
entropy_LSTM_tpr = entropy_LSTM.tpr
```

```python
baseline_MLP_auc = np.trapz(baseline_MLP_tpr, x=baseline_MLP_fpr)
baseline_LSTM_auc = np.trapz(baseline_LSTM_tpr, x=baseline_LSTM_fpr)

entropy_MLP_auc = np.trapz(entropy_MLP_tpr, x=entropy_MLP_fpr)
entropy_LSTM_auc = np.trapz(entropy_LSTM_tpr, x=entropy_LSTM_fpr)
```

```python
plt.figure(figsize=(5,4))
plt.plot(baseline_MLP_fpr, baseline_MLP_tpr, label="Baseline-CNN ROC AUC: {:.3f}".format(-baseline_MLP_auc),color = '#82B0D2',linewidth = 2)
plt.plot(baseline_LSTM_fpr, baseline_LSTM_tpr, label="Baseline-LSTM ROC AUC: {:.3f}".format(-baseline_LSTM_auc),color = '#8ECFC9',linewidth = 2)
plt.plot(entropy_MLP_fpr, entropy_MLP_tpr, label="Entropy-MLP ROC AUC: {:.3f}".format(-entropy_MLP_auc), color = '#FFBE7A',linewidth = 2)
plt.plot(entropy_LSTM_fpr, entropy_LSTM_tpr, label="Entropy-MLP ROC AUC: {:.3f}".format(-entropy_LSTM_auc), color = '#FA7F6F',linewidth = 2)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.rcParams.update({'font.size': 12})
plt.savefig('plot_result_ROC.png', bbox_inches="tight", dpi=700)
plt.show()
```
