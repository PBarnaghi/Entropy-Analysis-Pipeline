```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
result_baseline_lr = pd.read_csv('result_baseline_svm_timewindow_20.csv', usecols=['test_recall','test_f1','test_acc','cv_score'])
result_entropy_lr = pd.read_csv('result_entropy_svm_timewindow_20.csv', usecols=['test_recall','test_f1','test_acc','cv_score'])
```

```python
result_baseline_lr = pd.read_csv('result_baseline_lr_timewindow_20.csv', usecols=['test_recall','test_f1','test_acc','cv_score'])
result_entropy_lr = pd.read_csv('result_entropy_lr_timewindow_20.csv', usecols=['test_recall','test_f1','test_acc','cv_score'])

result_baseline_lr_test_recall = pd.DataFrame(result_baseline_lr['test_recall']).reset_index(drop=True)
result_baseline_lr_test_recall.columns = ['value']
result_baseline_lr_test_recall['evaluation'] = 'Recall rate'

result_baseline_lr_test_f1 = pd.DataFrame(result_baseline_lr['test_f1']).reset_index(drop=True)
result_baseline_lr_test_f1.columns = ['value']
result_baseline_lr_test_f1['evaluation'] = 'F1 score'

result_baseline_lr_test_acc = pd.DataFrame(result_baseline_lr['test_acc']).reset_index(drop=True)
result_baseline_lr_test_acc.columns = ['value']
result_baseline_lr_test_acc['evaluation'] = 'Accuracy'

result_baseline_lr_cv_score = pd.DataFrame(result_baseline_lr['cv_score']).reset_index(drop=True)
result_baseline_lr_cv_score.columns = ['value']
result_baseline_lr_cv_score['evaluation'] = 'CV score'

result_baseline_lr_all = pd.concat([result_baseline_lr_test_recall,result_baseline_lr_test_f1,result_baseline_lr_test_acc,result_baseline_lr_cv_score],axis=0)
result_baseline_lr_all['Group'] = 'Baseline'

result_entropy_lr_test_recall = pd.DataFrame(result_entropy_lr['test_recall']).reset_index(drop=True)
result_entropy_lr_test_recall.columns = ['value']
result_entropy_lr_test_recall['evaluation'] = 'Recall rate'

result_entropy_lr_test_f1 = pd.DataFrame(result_entropy_lr['test_f1']).reset_index(drop=True)
result_entropy_lr_test_f1.columns = ['value']
result_entropy_lr_test_f1['evaluation'] = 'F1 score'

result_entropy_lr_test_acc = pd.DataFrame(result_entropy_lr['test_acc']).reset_index(drop=True)
result_entropy_lr_test_acc.columns = ['value']
result_entropy_lr_test_acc['evaluation'] = 'Accuracy'

result_entropy_lr_cv_score = pd.DataFrame(result_entropy_lr['cv_score']).reset_index(drop=True)
result_entropy_lr_cv_score.columns = ['value']
result_entropy_lr_cv_score['evaluation'] = 'CV score'

result_entropy_lr_all = pd.concat([result_entropy_lr_test_recall,result_entropy_lr_test_f1,result_entropy_lr_test_acc,result_entropy_lr_cv_score],axis=0)
result_entropy_lr_all['Group'] = 'Entropy'

result_all = pd.concat([result_baseline_lr_all,result_entropy_lr_all],axis=0)
plt.figure(figsize=(8,5))
ax = sns.boxplot(x="evaluation", y="value", hue="Group",data=result_all, palette="magma")
plt.ylim(0,1)
plt.ylabel('Value')
plt.xlabel('Evaluation')
plt.rcParams.update({'font.size': 15})
plt.savefig('plot_result_LR.png', dpi=100)
```

```python
result_baseline_svm = pd.read_csv('result_baseline_svm_timewindow_20.csv', usecols=['test_recall','test_f1','test_acc','cv_score'])
result_entropy_svm = pd.read_csv('result_entropy_svm_timewindow_20.csv', usecols=['test_recall','test_f1','test_acc','cv_score'])

result_baseline_svm_test_recall = pd.DataFrame(result_baseline_svm['test_recall']).reset_index(drop=True)
result_baseline_svm_test_recall.columns = ['value']
result_baseline_svm_test_recall['evaluation'] = 'Recall rate'

result_baseline_svm_test_f1 = pd.DataFrame(result_baseline_svm['test_f1']).reset_index(drop=True)
result_baseline_svm_test_f1.columns = ['value']
result_baseline_svm_test_f1['evaluation'] = 'F1 score'

result_baseline_svm_test_acc = pd.DataFrame(result_baseline_svm['test_acc']).reset_index(drop=True)
result_baseline_svm_test_acc.columns = ['value']
result_baseline_svm_test_acc['evaluation'] = 'Accuracy'

result_baseline_svm_cv_score = pd.DataFrame(result_baseline_svm['cv_score']).reset_index(drop=True)
result_baseline_svm_cv_score.columns = ['value']
result_baseline_svm_cv_score['evaluation'] = 'CV score'

result_baseline_svm_all = pd.concat([result_baseline_svm_test_recall,result_baseline_svm_test_f1,result_baseline_svm_test_acc,result_baseline_svm_cv_score],axis=0)
result_baseline_svm_all['Group'] = 'Baseline'

result_entropy_svm_test_recall = pd.DataFrame(result_entropy_svm['test_recall']).reset_index(drop=True)
result_entropy_svm_test_recall.columns = ['value']
result_entropy_svm_test_recall['evaluation'] = 'Recall rate'

result_entropy_svm_test_f1 = pd.DataFrame(result_entropy_svm['test_f1']).reset_index(drop=True)
result_entropy_svm_test_f1.columns = ['value']
result_entropy_svm_test_f1['evaluation'] = 'F1 score'

result_entropy_svm_test_acc = pd.DataFrame(result_entropy_svm['test_acc']).reset_index(drop=True)
result_entropy_svm_test_acc.columns = ['value']
result_entropy_svm_test_acc['evaluation'] = 'Accuracy'

result_entropy_svm_cv_score = pd.DataFrame(result_entropy_svm['cv_score']).reset_index(drop=True)
result_entropy_svm_cv_score.columns = ['value']
result_entropy_svm_cv_score['evaluation'] = 'CV score'

result_entropy_svm_all = pd.concat([result_entropy_svm_test_recall,result_entropy_svm_test_f1,result_entropy_svm_test_acc,result_entropy_svm_cv_score],axis=0)
result_entropy_svm_all['Group'] = 'Entropy'

result_all = pd.concat([result_baseline_svm_all,result_entropy_svm_all],axis=0)
plt.figure(figsize=(8,5))
ax = sns.boxplot(x="evaluation", y="value", hue="Group",data=result_all, palette="magma")
plt.ylim(0,1)
plt.ylabel('Value')
plt.xlabel('Evaluation')
plt.rcParams.update({'font.size': 15})
plt.savefig('plot_result_SVM.png', dpi=100)
```
