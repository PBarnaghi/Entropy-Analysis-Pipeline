```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
old_1 = pd.DataFrame(np.loadtxt('o1-76-si.txt'), columns=['time','gait'])
young_1 = pd.DataFrame(np.loadtxt('y1-23-si.txt'), columns=['time','gait'])
parkinson_1 = pd.DataFrame(np.loadtxt('pd1-si.txt'), columns=['time','gait'])
```

```python
old_1
```

```python
plt.figure(figsize=(8, 4))
plt.plot(old_1.time, old_1.gait, color='#82b0d2')
plt.plot(young_1.time, young_1.gait, color = '#ffbe7a')
plt.plot(parkinson_1.time, parkinson_1.gait, color = '#fa7f6f')
plt.xlabel('Time/s', size=15)
plt.ylabel('Stride Intervals/s',size=15)
plt.legend(['Healthy older adults','Healthy young adults','Older adults with Parkinson\'s disease'], fontsize = 12)
plt.tick_params(labelsize=15)
plt.savefig('plot_original_data.png',dpi=750, bbox_inches = 'tight')
plt.show()
```
