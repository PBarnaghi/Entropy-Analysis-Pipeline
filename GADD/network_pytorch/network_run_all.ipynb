```python
import pandas as pd
import matplotlib.pyplot as plt
baseline_mlp = pd.read_csv('n_evaluation_baseline_mlp.csv') # 5 8 1
baseline_mlp.drop(['Unnamed: 0'],axis=1, inplace=True)
baseline_mlp.boxplot()
plt.ylim(0.5,1)
```

```python
baseline_lstm = pd.read_csv('n_evaluation_baseline_lstm.csv') # 5 16 1
baseline_lstm.drop(['Unnamed: 0'],axis=1, inplace=True)
baseline_lstm.boxplot()
plt.ylim(0.5,1)
```

```python
entropy_mlp = pd.read_csv('n_evaluation_entropy_mlp.csv') # 12 16 1
entropy_mlp.drop(['Unnamed: 0'],axis=1, inplace=True)
entropy_mlp.boxplot()
plt.ylim(0.5,1)
```

```python
entropy_lstm = pd.read_csv('n_evaluation_entropy_lstm.csv')
entropy_lstm.drop(['Unnamed: 0'],axis=1, inplace=True)
entropy_lstm.boxplot()
plt.ylim(0.5,1)
```

```python
print('baseline_mlp',baseline_mlp.mean())
print('baseline_lstm',baseline_lstm.mean())
print('entropy_mlp',entropy_mlp.mean())
print('entropy_lstm',entropy_lstm.mean())
```
